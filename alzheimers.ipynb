```python
print("missing values summary:")
print(df.isnull().sum())
```

9 out of 31 columns have missing values, which makes the dataset incomplete

```python
print("after removing irrelevant columns:")
df.drop(columns=['Data_Value_Alt', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote'], inplace=True)
print(df.isnull().sum())
```

some irrelevant columns like Data_Value_Alt, Data_Value_Footnote_Symbol, Data_Value_Footnote were dropped

```python
print("after filling data")
cols = ['Data_Value', 'Low_Confidence_Limit', 'High_Confidence_Limit']
df[cols] = df[cols].fillna(df[cols].mean())
df['StratificationCategory2'].fillna('unknown', inplace=True)
df['Stratification2'].fillna('unknown', inplace=True)
df['Geolocation'].fillna('unknown', inplace=True)
print(df.isnull().sum())
```

the columns data_value, low_confidence_limit, high_confidence_limit are necessary for analysis but cant be dropped entirely, so the missing datas are filled with the mean. Filling with mean ensures no extreme outliers. Furthermore, missing entries in StratificationCategory2, Stratification2, and Geolocation are filled with 'unknown'

```python
print("after replacing negative values in low_confidence_limit:")
df['Low_Confidence_Limit'] = df['Low_Confidence_Limit'].clip(lower=0)
print(df.describe())
```

the negative values in low_confidence_limit were replaced by 0

```python
df.rename(columns={'LocationAbbr': 'Location', 'LocationDesc': 'State'}, inplace=True)
```

for easier readability, the column names of 'LocationAbbr' and 'LocationDesc' were changed to 'Location' and 'State' respectively

```python
df = df.drop_duplicates()
```

```python
df['LocationID'] = df['LocationID'].astype(str)
```
