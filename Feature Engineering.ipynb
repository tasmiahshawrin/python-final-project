```python
# Feature 3: Data Value Normalized
df['Data_Value_Normalized'] = (
    (df['Data_Value'] - df['Data_Value'].min()) /
    (df['Data_Value'].max() - df['Data_Value'].min())
)
print("Feature 3: Data_Value_Normalized created")
print(df['Data_Value_Normalized'].describe())
```
