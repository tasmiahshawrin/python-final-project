```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Alzheimer_s_Disease_and_Healthy_Aging_Data.csv')
```

```python
df['Data_Value_Range'] = df['High_Confidence_Limit'] - df['Low_Confidence_Limit']
print("Feature 1: Data_Value_Range created")
print(df['Data_Value_Range'].describe())
```

```python
#  Grouping & Aggregation - Average Data_Value by Class 
print("Analysis 1: Average Data_Value by Class")
class_avg = df.groupby('Class')['Data_Value'].mean().sort_values(ascending=False).reset_index()
class_avg.columns = ['Class', 'Average_Data_Value']
print(class_avg)
```

```python
year_analysis = df.groupby('YearStart')['Data_Value'].mean()
print(year_analysis)
```

```python
state_analysis = df.groupby('LocationDesc')['Data_Value'].mean().sort_values(ascending=False)
print(state_analysis.head(10))
```

```python
category_analysis = df['Risk_Category'].value_counts()
print("Risk Category Distribution:")
print(category_analysis)
```

```python
# Step 1: Find the boundaries
Q1 = df['Data_Value'].quantile(0.25)
Q3 = df['Data_Value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
#  outlier analysis 
lower = df['Data_Value'].quantile(0.01)
upper = df['Data_Value'].quantile(0.99)

outliers = df[(df['Data_Value'] < lower) | (df['Data_Value'] > upper)]
print("Outliers using 1st and 99th percentile:")
print(f"Total Outliers Found: {len(outliers)}")
print(f"Outlier Percentage: {(len(outliers)/len(df))*100:.2f}%")
print(outliers[['Class', 'Topic', 'Data_Value', 'YearStart']].head(10))
```
